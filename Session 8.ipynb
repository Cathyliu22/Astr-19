### Trigonometric Function Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array x ranging from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# Define functions that return sin(x) and cos(x)
def sin_x(x):
    return np.sin(x)

def cos_x(x):
    return np.cos(x)
```

```python
# Create a multipanel plot (1 row, 2 columns)
f,axarr = plt.subplots(1, 2, figsize=(10, 5))

# Plot sin(x) vs. x in the left panel
axarr[0].plot(x, sin_x(x))
axarr[0].set_title(r'$\sin(x)$')
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')

# Plot cos(x) vs. x in the right panel
axarr[1].plot(x, cos_x(x))
axarr[1].set_title(r'$\cos(x)$')
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
```

```python
# Save the figure as a PDF
plt.savefig('trig_plot.pdf')
plt.show()
```
