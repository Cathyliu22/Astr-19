### Exponential Function Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array x ranging from 0 to 1 in 100 steps
x = np.linspace(0, 1, 100)
```

```python
# Define a function that returns exp(x)
def exp_x(x):
    return np.exp(x)
```

```python
# Set y = exp(x)
y = exp_x(x)
```

```python
# Plot x vs. y
plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title('Exponential Function')
plt.xlabel('Time [milliseconds]')
plt.ylabel('Awesomeness')
```

```python
# Save the figure as a PDF
plt.savefig('exp_plot.pdf')
plt.show()
```
