### Histogram of Random Numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the random seed for reproducibility
np.random.seed(42)
```

```python
# Generate 1000 random numbers from a normal distribution (mean=0, std=1)
data = np.random.normal(loc=0, scale=1, size=1000)
```

```python
# Create a histogram of the data
plt.figure(figsize=(12, 5))
plt.hist(data, bins=100, color='green', edgecolor='black')
plt.title('Histogram of 1000 Random Numbers from Normal Distribution')
plt.xlabel('value')
plt.ylabel('frequency')
```

```python
# Save the figure as a PDF
plt.savefig('normal_distribution_histogram.pdf')
plt.show()
```
