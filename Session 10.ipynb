### Histogram of Random Numbers

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate 1000 random numbers uniformly distributed between [0, 1]
random_numbers = np.random.uniform(0, 1, 1000)
```

```python
# Create a histogram of the random numbers with 100 bins
plt.figure(figsize=(12, 5))
plt.hist(random_numbers, bins=100, color='green', edgecolor='black')
plt.title('Histogram of Random Numbers')
plt.xlabel('value')
plt.ylabel('frequency')
```

```python
# Save the figure as a PDF
plt.savefig('random_numbers_histogram.pdf')
plt.show()
```
